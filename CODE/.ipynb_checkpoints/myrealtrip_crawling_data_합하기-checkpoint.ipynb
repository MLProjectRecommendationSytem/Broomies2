# 상품소개 글에서 특정 문자 제거

```python
def del_words(content):
    remove_words = ["\r\n\r\n1.", "\r\n", "\r\n\r\n**", "\r\n\r\n", "\r\n3", "-", "★", "■", "•", "↓", "=", "/", "※"]
    for remove_word in remove_words:
        content = content.replace(remove_word, "")
    return content
```

```python
data_2["intro"] = data_2["introduction"].apply(del_words)
```

```python
data_2["intro"][197]
```

# 평점 반올림

```python
data_2["score"] = round(data_2["score"], 2)
```

```python
data_2[:3]
```

# travel_purpose 의 키 값 분리

```python
data_2["purpose"] = data_2["travel_purpose"].apply(lambda data : data["purpose"])
data_2["age_group"] = data_2["travel_purpose"].apply(lambda data : data["age_group"])
```

# 컬럼 정리

```python
data_2 = data_2[["code", "score", "selling_count", "intro",  "reviews_count", "travel_purpose"]]
data_2
```

# eval 코드를 for 문에 넣어서 str 안에 dic를 꺼낸다.

```python
dict = []
for idx in range(0, 555):
    str = code_3.iloc[idx, 1]
    dict.append(eval(str))
```

```python
purpose_age = pd.DataFrame(dict)
purpose_age
```

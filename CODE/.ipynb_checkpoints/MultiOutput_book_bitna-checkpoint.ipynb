```python
temp = []
temp.append(list(data['interest'].unique()))
temp.append(list(data['area'].unique()))
temp.append(list(data['company'].unique()))
temp = sum(temp, [])
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(temp)
```

```python
le.classes_
```

```python
le.transform(data['interest'])
```

```python
data['y1'] = data['area'].apply(lambda x: le.transform([x])[0])
data['y2'] = data['company'].apply(lambda x : le.transform([x])[0])
data['y3'] = data['interest'].apply(lambda x : le.transform([x])[0])
```

```python
data['new_keyword'][0]
```

```python
data = data.fillna(' ')
```

```python
data.head()
```

```python
data['temp'] = data['title'] + data['new_keyword'] + data['new_subject'] + data['author'] + data['publisher']
data
```

### Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier

est = [('tf', TfidfVectorizer())]
pipe = Pipeline(est)
```

```python
from sklearn.model_selection import train_test_split

X = pipe.fit_transform(data['temp']).toarray()
Y = np.vstack((np.array(data['y1']),np.array(data['y2']),np.array(data['y3']))).T

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=4)
```

```python
X_train
```

```python
forest = RandomForestClassifier(random_state=2, max_features=60, n_estimators=40)
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)
model = multi_target_forest.fit(X_train, y_train)
predicted = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
